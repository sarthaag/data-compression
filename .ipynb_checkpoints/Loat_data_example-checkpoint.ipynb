```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# load data
All_phi = np.load("Phi_201snap.npy")
R0 = 1.7853     # (m) major radius
xx = np.load('meshR.npy')*R0
zz = np.load('meshZ.npy')*R0
All_phi.shape
```

```python
# visualize modes
fig = plt.figure(figsize = (6, 6))
plt.pcolor(xx, zz, All_phi[:,:,120], shading='auto', cmap='seismic')
plt.axis('scaled')
plt.colorbar(orientation='horizontal', shrink=.5)
plt.xlabel(r'$X/R_0$')
plt.ylabel(r'$Z/R_0$')
# plt.clim([-1,1])
plt.show()
```
