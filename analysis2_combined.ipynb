```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import all data and combine them together
df1 = pd.read_pickle("analysis2.1.pkl")
df2 = pd.read_pickle("analysis2.2.pkl")
df3 = pd.read_pickle("analysis2.3.pkl")

df = pd.concat([df1, df2], axis=1)

# Create X and Y lists to use for pyplot
X = [0, df3.at[0, 'RMSE32'], df3.at[0, 'RMSE16']]
Y = [1, 2, 4]
df
```

```python
# Analysis 2.3

plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color='red')
plt.plot(X, Y, color='red')
plt.xscale("log")
plt.xlabel("RMSE")
plt.ylabel("Compression Ratio")
plt.title("Storing Arrays to FP32 and FP16")
plt.show()
```

```python
# Analysis 2.1

plt.plot(df['RMSE2.1'], df['CR2.1'], color='blue')
plt.xlabel("RMSE")
plt.ylabel("Compression Ratio")
plt.xscale("log")
plt.title("Plotting Arrays as 101x227 Grayscale")
plt.show()
```

```python
# Analysis 2.2

plt.plot(df['RMSE2.2'], df['CR2.2'], color='green')
plt.xlabel("RMSE")
plt.ylabel("Compression Ratio")
plt.xscale("log")
plt.title("Plotting Arrays as 203x203 Grayscale")
plt.show()
```

```python
# Combined graphs

plt.scatter(X, Y, color='red', label='Graph 1')
plt.plot(X, Y, color='red', label='Graph 1')
plt.plot(df['RMSE2.1'], df['CR2.1'], color='blue', label='Graph 2')
plt.plot(df['RMSE2.2'], df['CR2.2'], color='green', label='Graph 3')
plt.legend()
plt.xlabel("RMSE")
plt.ylabel("Compression Ratio")
plt.title("Combined Graph")
plt.xscale("log")
plt.show()
```

```python
# Plotting all four graphs together

# Create 3 plots for each analysis and 1 combined plot
figure, axis = plt.subplots(2, 2)

axis[0, 0].plot(X, Y)
# axis[0, 0].set_title("Storing Arrays to FP32 and FP16")

axis[0, 1].plot(df['norm2.1'], df['CR2.1'])
# axis[0, 1].set_title("Plotting Arrays as 101x227 Grayscale")

axis[1, 0].plot(df['norm2.2'], df['CR2.2'])
# axis[1, 0].set_title("Plotting Arrays as 203x203 Grayscale")

axis[1, 1].plot(X, Y, label='1')
axis[1, 1].plot(df['norm2.1'], df['CR2.1'], label='2')
axis[1, 1].plot(df['norm2.2'], df['CR2.2'], label='3')
axis[1, 1].legend()
# axis[1, 1].set_title("Combined Plot")

# plt.xticks([3.25,3.5,3.75,4,4.25])
plt.show()
```
