```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import all data
df1 = pd.read_pickle("analysis2.1.pkl")
df2 = pd.read_pickle("analysis2.2.pkl")
df3 = pd.read_pickle("analysis2.3.pkl")
df4 = pd.read_pickle("analysis2.4.pkl")

df = pd.concat([df1, df2], axis=1)

# Create X and Y lists to use for pyplot
X = [df3.at[0, 'RMSE32'], df3.at[0, 'RMSE16']]
Y = [2, 4]
```

```python
# Analysis 2.3

plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color='red')
plt.xlabel("RMSE")
plt.ylabel("Compression Ratio")
plt.title("Storing Arrays to FP32 and FP16")
plt.show()
```

```python
# Analysis 2.1

plt.figure(figsize=(10, 6))
plt.scatter(df['RMSE2.1.95'], df['CR2.1.95'], color='red', label='95', marker='.')
plt.scatter(df['RMSE2.1.75'], df['CR2.1.75'], color='green', label='75', marker='.')
plt.scatter(df['RMSE2.1.50'], df['CR2.1.50'], color='blue', label='50', marker='.')
plt.scatter(df['RMSE2.1.25'], df['CR2.1.25'], color='purple', label='25', marker='.')
plt.legend(title='% Quality')

# Plotting the mean for each quality
plt.scatter(df['RMSE2.1.95'].mean(axis=0), df['CR2.1.95'].mean(axis=0), color='black', marker='D')
plt.scatter(df['RMSE2.1.75'].mean(axis=0), df['CR2.1.75'].mean(axis=0), color='black', marker='D')
plt.scatter(df['RMSE2.1.50'].mean(axis=0), df['CR2.1.50'].mean(axis=0), color='black', marker='D')
plt.scatter(df['RMSE2.1.25'].mean(axis=0), df['CR2.1.25'].mean(axis=0), color='black', marker='D')

plt.xlabel("RMSE")
plt.ylabel("Compression Ratio")
plt.title("Plotting Arrays as 101x227 Grayscale")
plt.show()
```
