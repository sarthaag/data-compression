```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import all data and combine them together
df1 = pd.read_pickle("analysis2.1.pkl")
df2 = pd.read_pickle("analysis2.2.pkl")
df3 = pd.read_pickle("analysis2.3.pkl")

df = pd.concat([df1, df2], axis=1)

# Create X and Y lists to use for pyplot
X = [0, df3.at[0, 'RMSE32'], df3.at[0, 'RMSE16']]
Y = [1, 2, 4]
df
```

```python
# Analysis 2.3

plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color='red')
plt.plot(X, Y, color='red')
plt.xscale("log")
plt.xlabel("RMSE")
plt.ylabel("Compression Ratio")
plt.title("Storing Arrays to FP32 and FP16")
plt.show()
```

```python
# Analysis 2.1

plt.figure(figsize=(10, 6))
plt.plot(df['RMSE2.1.100'], df['CR2.1.100'], color='red', label='100')
plt.plot(df['RMSE2.1.75'], df['CR2.1.75'], color='green', label='75')
plt.plot(df['RMSE2.1.50'], df['CR2.1.50'], color='blue', label='50')
plt.plot(df['RMSE2.1.25'], df['CR2.1.25'], color='purple', label='25')
plt.legend(title='% Quality')

plt.xlabel("RMSE")
plt.ylabel("Compression Ratio")
plt.xscale("log")
plt.title("Plotting Arrays as 101x227 Grayscale")
plt.show()
```

```python
# Analysis 2.2

plt.figure(figsize=(10, 6))
plt.plot(df['RMSE2.2.100'], df['CR2.2.100'], color='red', label='100')
plt.plot(df['RMSE2.2.75'], df['CR2.2.75'], color='green', label='75')
plt.plot(df['RMSE2.2.50'], df['CR2.2.50'], color='blue', label='50')
plt.plot(df['RMSE2.2.25'], df['CR2.2.25'], color='purple', label='25')
plt.legend(title='% Quality')

plt.xlabel("RMSE")
plt.ylabel("Compression Ratio")
plt.xscale("log")
plt.title("Plotting Arrays as 203x203 Grayscale")
plt.show()
```

```python
# Combined graphs

plt.figure(figsize=(10, 6))
plt.xscale("log")
plt.scatter(X, Y, color='red', label='Graph 1')
plt.plot(X, Y, color='red', label='Graph 1')
plt.plot(df['RMSE2.1'], df['CR2.1'], color='blue', label='Graph 2')
plt.plot(df['RMSE2.2'], df['CR2.2'], color='green', label='Graph 3')
plt.legend(loc='lower right')
plt.xlabel("RMSE")
plt.ylabel("Compression Ratio")
plt.title("Combined Graph")
plt.show()
```
